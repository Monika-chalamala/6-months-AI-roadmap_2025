```python
# comparison of two players in one graph.

plt.plot(Salary[0], color='green', ls = ':', marker='*', label= Players[0], ms= 7)
plt.plot(Salary[1], color='purple', ls = '--', marker='o', label= Players[1], ms = 7)

plt.xticks(list(range(0,10)), Seasons, rotation='vertical')

plt.show()
```

```python
# the legend() function is used for labeling or to create the index box.

plt.plot(Salary[0], color='green', ls = ':', marker='*', label= Players[0], ms= 7)
plt.plot(Salary[1], color='purple', ls = '--', marker='o', label= Players[1], ms = 7)
plt.plot(Salary[2], color='cyan', ls = '-', marker='s', label= Players[2], ms= 7)
plt.plot(Salary[3], color='magenta', ls = '-.', marker='x', label= Players[3], ms= 7)

plt.legend()
plt.xticks(list(range(0,10)), Seasons, rotation='vertical')

plt.show()
```
