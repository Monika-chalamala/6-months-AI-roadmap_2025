```python
# importing pyplot as plt from matplotlib
import matplotlib.pyplot as plt
```

```python
# importing Image from Python Imaging Library
from PIL import Image
```

```python
# open() functions opens the file at the loaction mentioned in argument
horse_image= Image.open(r'C:\Users\DELL\Desktop\horse.webp')
```

```python
horse_image
```

```python
# checking the datatype
type(horse_image)
# PIL means Python Imaging Library which handles one of the unstructured data i.e; images.
```

```python
# creating a variable and converting image into ndmatrix or ndarray
horse_arr=np.asarray(horse_image)
```

```python
horse_arr
```

```python
# checking the datatype
type(horse_arr)
```

```python
# imshow() function shows the image
plt.imshow(horse_arr)
```

```python
# checking the dimensions
horse_arr.shape
```

```python
# copying array 
horse_red = horse_arr.copy()
```

```python
horse_red
```

```python
# ensuring whether the data is copied properly 
horse_arr == horse_red
```

```python
# imshow() function shows the image
plt.imshow(horse_red)
```

```python
# the 0 here indicates the index in particular channel
plt.imshow(horse_red[:,:,0])
```

```python
horse_red[:,:,0]
```

```python
# we are using greys scale on the cmap and 0 indicates the index of the channel.
plt.imshow(horse_red[:,:,0], cmap = 'Greys') 
```

```python
# we are using grayscale on the cmap and 1 indicates the index of the channel.
plt.imshow(horse_red[:,:,1], cmap = 'grey')
```

```python
# we are using grayscale on the cmap and 0 indicates the index of the channel.
plt.imshow(horse_red[:,:,0], cmap = 'grey')
```

```python
# we changed the color in cmap YlGn means Yellow Green , this shows the intensity of yellow and green in the picture.
plt.imshow(horse_red[:,:,1], cmap = 'YlGn')
```
