```python
# gtts is the library for text to speech conversion
from gtts import gTTS
from IPython.display import Audio

text_to_speech =gTTS('''Hi there, my name is Monika Chalamala, a Btech Graduate in Data Science specialization.
                     As my Degree taught me basics and formed a foundation for my data science career, I choose 
                     Full Stack Data Science and AI by Naresh I Technologies course to build a strong skillset 
                     to elevate my profile.''', lang='hi', tld='com')

text_to_speech.save('text_to_speech_gtts.wav')
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file, autoplay= False)
```

# CHANGING GENDER OF THE AUDIO

```python
!pip install pyttsx3
```

```python
import pyttsx3
from IPython.display import Audio

text = ('''Hi there, my name is Monika Chalamala, a Btech Graduate in Data Science specialization.
                     As my Degree taught me basics and formed a foundation for my data science career, I choose 
                     Full Stack Data Science and AI by Naresh I Technologies course to build a strong skillset 
                     to elevate my profile.''')

audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)

#change the voices
voice = audio.getProperty('voices')

# 0 for male and 1 for female
#audio.setProperty('voice', voice[0].id)
audio.setProperty('voice', voice[1].id)

# text to speech conversion
audio.say(text)

#save the audio file
#audio.save_to_file(text, 'test_male_voice.mp3')
audio.save_to_file(text, 'test_female_voice.mp3')

audio.runAndWait()
```
